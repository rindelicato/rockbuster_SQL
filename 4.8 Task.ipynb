```python
#import libraries
import pandas as pd
import numpy as np
import os
```

```python
path = r'C:\Users\rinde\Downloads\Instacart-Basket-Analysis-11-2023\Instacart Basket Analysis 11-2023'
```

```python
ords_prods_merged = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl'))
```

```python
# 2 Aggregating mean of order_number grouped department_id for entire dataframe
```

```python
#Checking dataframe
ords_prods_merged.head()
```

```python
#perform single aggregation mean of order number by department_id (average number of orders per user department id)
ords_prods_merged.groupby('department_id')['order_number'].mean()
```

```python
#3 Analyze result
# There are 21 department_id's compared to the subset of 100000 there are only 8
```

```python
#4 Creating flags for loyalty customers
```

```python
# Creating new column "max_order"
ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking for new column
ords_prods_merged.head(20)
```

```python
# Assign loyalty flag
ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check loyalty flags
ords_prods_merged['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check Loyalty flag in dataframe
ords_prods_merged[['user_id', 'loyalty_flag', 'order_number']].head(20)
```

```python
#5 Customers spending habits comparison
```

```python
# the min max and mean of each loyalty flag
ords_prods_merged.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
# loyal customers spend the least, new customers spend the most, and regular cusotmers fall in between. This is seen by observing the mean.
```

```python
#6 Creating Flag on price mean of Spender
```

```python
# Creating column for average spender
ords_prods_merged['average_spend'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking the new column
ords_prods_merged.head(20)
```

```python
# Creating flag for low and high spender
ords_prods_merged.loc[ords_prods_merged['average_spend']<10,'spender_flag']='Low_spender'
```

```python
ords_prods_merged.loc[ords_prods_merged['average_spend']>= 10, 'spender_flag'] = 'High_spender'
```

```python
# Check spender flags
ords_prods_merged['spender_flag'].value_counts(dropna = False)
```

```python
ords_prods_merged[['user_id', 'average_spend', 'spender_flag', 'prices']].head(60)
```

```python
#7 Create a flag for frequent users
```

```python
# Create a column for frequet users
ords_prods_merged['Customer_frequency'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Cheaking for new column
ords_prods_merged.head(20)
```

```python
# Creating flag on order frequency
ords_prods_merged.loc[ords_prods_merged['Customer_frequency'] > 20, 'frequency_flag'] = 'Non_frequent_customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['Customer_frequency']>10) & (ords_prods_merged['Customer_frequency']<=20),
                     'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['Customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent Customer'
```

```python
# Check frequency flags
ords_prods_merged['frequency_flag'].value_counts(dropna = False)
```

```python
#Checking for columns
ords_prods_merged[['user_id', 'Customer_frequency', 'frequency_flag']].head(60)
```
